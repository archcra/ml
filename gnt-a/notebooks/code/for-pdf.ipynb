
关于测序实验中，对DNA测序重数的研究
=====

# 前言
一般在考虑DNA测序重数时，都认为样品是纯样。而在植物的DNA取样上，混样更常见。对于混样，测序的DNA重数取多少合适？不同的混样比例，是否重数还需要变化？对于杂和基因型的情况呢？

# 问题的提出
将样品的DNA扩增后，从中抽取一定数量的DNA，进行测序，如取2000条。 目前扩增可以达到10万条。 但是，取多少条，能准确代表原来10万条的DNA？ 因为取的越多，则测的越多，即所谓的重数越多，成本越高。 如果是纯样，没有混样，则可能测10重就够了。 如果是混样，则情况可能完全不同。

现在的问题，假设为混样的情况。混样的主样品基因型占比分别可能占90%，60%, 50%及30%。试分析取多少重，能达到相对准确地代表原基因的占比信息？

程序模拟： 设有10万个数字，由0和1组成。用1的占比来模拟主样品基因型的占比。如模拟主样品基因型占比为90%，时，则将数据中，设置1的含量为90%，随机分布。
现在的问题就变成，当1的占比分别为90%，60%, 50%及30%时,取多少组，其占比率的变化，都基本和原来一致？即，如取1000条，则在取之后，1在这1000条中， 也是分别占90%，60%, 50%和30%左右？

# 解决思路
我们使用蒙特卡罗大量采样法，来计算不同重数时，取样后的占比误差。

以占比为90%为例，具体步骤如下：
A. 生成10万个数字，里面包含90%的随机概率的1，其余数字为0。
B. 然后，从中取机取出100个，计算数字1的占比；
重复B步骤多次，如1000次或10000万次，然后计算占比均方差。这个重复次数越多，生成的占比均方差的跳动越小，即越接近真值。
C. 重复B，将取出的个数增加1。计算均方差。
D. 重复C，直致每次取出2000个，分别计算均方差。
根据B，C，D步骤的均方差结果绘图。其中，横轴为取出数字的个数，纵轴为均方差。

分别使用概率60%, 50%, 30%来重复上面A～D步骤，做出另外3种概率的图。
